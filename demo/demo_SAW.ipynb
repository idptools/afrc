# `SAW` demo

The `SAW` class implements a self-avoiding walk model. Its dimensions are tuned by a `prefactor` argument (default 5.5 A).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from afrc.polymer_models.saw import SAW

seq = 'MASNDYTQQATQSYGAYPTQPGQGYSQQSSQPYGQQSYSGYSQSTDTSGYG'
model = SAW(seq)
```

```python
print(f"mean Re : {model.get_mean_end_to_end_distance():.2f} A")
print(f"RMS Re  : {model.get_root_mean_squared_end_to_end_distance():.2f} A")
print(f"mean Rg : {model.get_mean_radius_of_gyration():.2f} A")
```

## Varying the prefactor

```python
plt.figure(figsize=(6, 3))
for pref in [4.5, 5.5, 6.5]:
    r, p = model.get_end_to_end_distribution(prefactor=pref)
    plt.plot(r, p, label=f'prefactor = {pref}')
plt.xlabel('Re (A)')
plt.ylabel('probability')
plt.legend()
plt.show()
```
