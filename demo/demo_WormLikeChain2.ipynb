# `WormLikeChain2` demo

The `WormLikeChain2` class implements the O'Brien et al. (2009) worm-like chain model. Unlike the Zhou model it also provides an estimate of the mean radius of gyration.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from afrc.polymer_models.wlc2 import WormLikeChain2

seq = 'MASNDYTQQATQSYGAYPTQPGQGYSQQSSQPYGQQSYSGYSQSTDTSGYG'
model = WormLikeChain2(seq)
```

```python
print(f"mean Re : {model.get_mean_end_to_end_distance():.2f} A")
print(f"RMS Re  : {model.get_root_mean_squared_end_to_end_distance():.2f} A")
print(f"mean Rg : {model.get_mean_radius_of_gyration():.2f} A")
```

## Varying the persistence length

```python
plt.figure(figsize=(6, 3))
for lp in [2.0, 3.0, 4.0]:
    r, p = WormLikeChain2(seq, lp=lp).get_end_to_end_distribution()
    plt.plot(r, p, label=f'lp = {lp} A')
plt.xlabel('Re (A)')
plt.ylabel('probability')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 3))

rg_vals = []
lp_vals = np.arange(1,20)
for lp in lp_vals:
    rg_vals.append(WormLikeChain2(seq, lp=lp).get_mean_radius_of_gyration())


plt.plot(lp_vals, rg_vals,'o')    
plt.xlabel('Persistence length (Lp)')
plt.ylabel('Radius of gyration')
plt.legend()
plt.show()
```
