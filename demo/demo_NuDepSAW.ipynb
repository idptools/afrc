# `NuDepSAW` demo

The `NuDepSAW` class implements a nu-dependent self-avoiding walk. The Flory scaling exponent `nu` can be varied continuously (nu=0.5 recovers ideal-chain dimensions, nu~0.6 a good solvent).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from afrc.polymer_models.nudep_saw import NuDepSAW

seq = 'MASNDYTQQATQSYGAYPTQPGQGYSQQSSQPYGQQSYSGYSQSTDTSGYG'
model = NuDepSAW(seq)
```

```python
for nu in [0.40, 0.50, 0.59]:
    re = model.get_mean_end_to_end_distance(nu=nu)
    rg = model.get_mean_radius_of_gyration(nu=nu)
    print(f'nu = {nu:.2f}   mean Re = {re:6.2f} A   mean Rg = {rg:6.2f} A')
```

## Varying the scaling exponent

```python
plt.figure(figsize=(6, 3))
for nu in [0.40, 0.50, 0.59]:
    r, p = model.get_end_to_end_distribution(nu=nu)
    plt.plot(r, p, label=f'nu = {nu:.2f}')
plt.xlabel('Re (A)')
plt.ylabel('probability')
plt.legend()
plt.show()
```
