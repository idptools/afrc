# `WormLikeChain` demo

The `WormLikeChain` class implements the Zhou (2004) worm-like chain model. It is a composition-independent reference model parameterised by a persistence length `lp` and a segment size `aa_size`.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from afrc.polymer_models.wlc import WormLikeChain

seq = 'MASNDYTQQATQSYGAYPTQPGQGYSQQSSQPYGQQSYSGYSQSTDTSGYG'
model = WormLikeChain(seq)
```

```python
print(f"mean Re      : {model.get_mean_end_to_end_distance():.2f} A")
print(f"RMS Re       : {model.get_root_mean_squared_end_to_end_distance():.2f} A")
```

## Varying the persistence length

Stiffer chains (larger `lp`) are more extended.

```python
plt.figure(figsize=(6, 3))
for lp in [2.0, 3.0, 4.0]:
    r, p = WormLikeChain(seq, lp=lp).get_end_to_end_distribution()
    plt.plot(r, p, label=f'lp = {lp} A')
plt.xlabel('Re (A)')
plt.ylabel('probability')
plt.legend()
plt.show()
```
